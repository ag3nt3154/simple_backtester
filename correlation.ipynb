```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
t1 = 'GLD'
t2 = 'SLV'
df1 = pd.read_csv(t1 + '_rec.csv')
df2 = pd.read_csv(t2 + '_rec.csv')


df = pd.merge(df1[['date', 'returns']], df2[['date', 'returns']], on='date')
df = df.set_index('date')
print(f"correlation is: {df['returns_x'].corr(df['returns_y'])}")
df['returns_x'].plot()
df['returns_y'].plot()
plt.xticks(rotation = 45) # Rotates X-Axis Ticks by 45-degrees
plt.show()
```
